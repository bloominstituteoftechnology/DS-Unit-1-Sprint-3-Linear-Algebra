## 1.8 Find $||g||$ and then find $||h||$. Which is longer?

\begin{align}
\vec{g} = \begin{bmatrix} 1 \\ 1 \\ 1 \\ 8 \end{bmatrix}
\qquad
\vec{h} = \begin{bmatrix} 3 \\3 \\ 3 \\ 3 \end{bmatrix}
\end{align}

```
g = np.array([[1],[1],[1],[8]])
h = np.array([[3],[3],[3],[3]])

print(g)
print(h)
```

```
norm_g = (g[0]**2 + g[1]**2 + g[2]**2 + g[3]**2)**(1/2)
norm_h = (h[0]**2 + h[1]**2 + h[2]**2 + h[3]**2)**(1/2)

print(norm_g)
print(norm_h)
```

```
#check

norm_g1 = np.linalg.norm(g)
norm_h1 = np.linalg.norm(h)

print(norm_g1)
print(norm_h1)
```
