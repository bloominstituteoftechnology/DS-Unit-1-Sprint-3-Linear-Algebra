## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```
# We're going to calculate the determinants of matrices H and J.

h = np.array([[12, 11], [7, 10]])
j = np.array([[0, 1, 2], [7, 10, 4], [3, 2, 0]])

round(np.linalg.det(h))
```

```
round(np.linalg.det(j))
```

## 3.3 Find $H^{-1}$ and then find $J^{-1}$

```
# We're going to calculate the inverse of matrices H and J.
inverse_h = np.linalg.inv(h)

inverse_h
```

```
inverse_j = np.linalg.inv(j)

inverse_j
```

Please ignore Python rounding errors. If necessary, format your output so that it rounds to 5 significant digits (the fifth decimal place).
