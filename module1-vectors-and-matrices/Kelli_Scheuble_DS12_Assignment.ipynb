## 1.3 Scale the vectors you created in 1.1 by $5$, $\pi$, and $-e$ and plot all four vectors (original + 3 scaled vectors) on a graph. What do you notice about these vectors? 

```
from math import e, pi
print(e)
print(pi)
```

```
# 1.1 is yellow
# 5 is orange
# pi is red
# -e is green

scalars = [1, 1.1, 5, pi, -e]

axes = plt.gca()
axes.set_xlim([-1, 1.5])
axes.set_ylim([-1.5, 1.75])

for scalar in scalars:
  if scalar == 1.1:
    scalar_color = 'y'
  elif scalar == 5:
    scalar_color = '#FF7F00'
  elif scalar == pi:
    scalar_color = 'r'
  elif scalar == -e:
    scalar_color = 'g'
  else:
    scalar_color = 'b'

  plt.arrow(0, 0, (.2 * scalar), (.3 * scalar), 
            head_width=.2, head_length=.2,
            color = scalar_color);


# The resulting graph is one with many different vectors on the same plane.
# The -e scalar flipped the direction, but scaled it to a length similar to that of
# the red vector, which was made with the scalar pi. The orange vector is the longest
# because that scalar was the largest with the value of 5.
```

```
vectors = []
three_d_vector = np.array([0,0,0,.2,.3,.2])

for scalar in scalars:
  vectors.append(three_d_vector * int(scalar))

X, Y, Z, U, V, W = zip(*vectors)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(X, Y, Z, U, V, W, length=1)
ax.set_xlim([0, 1.5])
ax.set_ylim([-1, 1.5])
ax.set_zlim([0, 1.5])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()

# pretty much the same thing here, but on 3D graph!
```

## 1.8 Find $||g||$ and then find $||h||$. Which is longer?

\begin{align}
\vec{g} = \begin{bmatrix} 1 \\ 1 \\ 1 \\ 8 \end{bmatrix}
\qquad
\vec{h} = \begin{bmatrix} 3 \\3 \\ 3 \\ 3 \end{bmatrix}
\end{align}

```
g = np.array([1,1,1,8])
h = np.array([3,3,3,3])
longer = []

for x in range(2):
  if x == 0:
    vector_name = 'a'
    vector = g
  if x == 1:
    vector = h
    vector_name = 'b'
  
  norm = np.linalg.norm(vector)
  longer.append(norm)
  print('norm of vector {} is {}'.format(vector_name,norm))

if longer[0] > longer[1]:
  print('vector a is longer than vector b')
else:
  print('vector b is longer than vector a')
# vector a is longer with a norm of 8.185
```


## 2.2 Find the following products: CD, AE, and BA. What are the dimensions of the resulting matrices? How does that relate to the dimensions of their factor matrices?

```
# C & D

def multiply(matrix1, matrix2):
  for i, x in enumerate ([matrix1.shape, matrix2.shape]):
    print('shape of matrix {} is {}:'.format(i+1, x))
  multiplied = np.matmul(matrix1,matrix2)
  print('shape of multiplied matrix:', multiplied.shape)
  print('resulting matrix:')
  print(multiplied)
  print()

print('C & D')
multiply(C,D)
print('A & E')
multiply(A,E)
print('B & A')
multiply(B,A)

# The dimensions of the resulting matrices are the dimensions of the 
# rows of the first matrix and the columns of the second matrix
```
