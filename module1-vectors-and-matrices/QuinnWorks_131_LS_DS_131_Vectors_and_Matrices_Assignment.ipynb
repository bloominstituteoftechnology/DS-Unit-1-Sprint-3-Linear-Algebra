## 2.2 Find the following products: CD, AE, and BA. What are the dimensions of the resulting matrices? How does that relate to the dimensions of their factor matrices?

```
CD = np.matmul(C,D)
AE = np.matmul(A,E)
BA = np.matmul(B,A)

print("C:",C.shape,"D:",D.shape)
print("CD:",CD.shape)
print(CD)


print("A:",A.shape,"E:",E.shape)
print("AE:",AE.shape)
print(AE)


print("B:",B.shape,"A:",A.shape)
print("BA:",BA.shape)
print(BA)
```

## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```
H = np.array([[12,11],
             [7,10]])
J = np.array([[0,1,2],
             [7,10,4],
             [3,2,0]])
```

```
#H Determinant
np.linalg.det(H)
```

```
#J Determinant
np.linalg.det(J)
```

```
#J Inverse
JI = np.linalg.inv(J)
JI 
```
