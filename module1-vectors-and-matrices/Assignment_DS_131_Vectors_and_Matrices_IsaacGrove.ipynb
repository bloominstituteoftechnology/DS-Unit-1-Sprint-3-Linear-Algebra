## 1.1 Create a two-dimensional vector and plot it on a graph

```
import matplotlib.pyplot as plt

my_2d_vector = [1.5, 2.5]

plt.arrow(0,0, my_2d_vector[0], my_2d_vector[1], head_width=0.06, head_length=0.06, color='cyan')
plt.title("My first vector")
plt.xlim(0,1.6)
plt.ylim(0,2.6)
plt.show()

```

## 1.2 Create a three-dimensional vector and plot it on a graph

```
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

my_3d_vector = [-0.2, 0.8, 0.7]

vector = np.array([[0, 0, 0, my_3d_vector[0], my_3d_vector[1], my_3d_vector[2]]])

X, Y, Z, U, V, W = zip(*vector)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(X, Y, Z, U, V, W, length=1)
ax.set_xlim([-0.5, 0.5])
ax.set_ylim([0, 1])
ax.set_zlim([0, 1])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

## 1.3 Scale the vectors you created in 1.1 by $5$, $\pi$, and $-e$ and plot all four vectors (original + 3 scaled vectors) on a graph. What do you notice about these vectors? 

```
from math import e, pi
print(e)
print(pi)

scaled_5 = np.multiply(5, my_2d_vector)
scaled_pi = np.multiply(pi, my_2d_vector)
scaled_e = np.multiply(-e, my_2d_vector)

plt.arrow(0,0, scaled_5[0], scaled_5[1], head_width=0.4, head_length=0.4, color='red')
plt.arrow(0,0, scaled_pi[0], scaled_pi[1], head_width=0.4, head_length=0.4, color='black')
plt.arrow(0,0, scaled_e[0], scaled_e[1], head_width=0.4, head_length=0.4, color='green')
plt.arrow(0,0, my_2d_vector[0], my_2d_vector[1], head_width=0.4, head_length=0.4, color='cyan')
plt.title("My first vector scaled by 5, pi, -e")
plt.xlim(-6,10)
plt.ylim(-10,15)
plt.show()

# They all have the same origin at (0,0), and the same slope...
# They're all part of the same line if you extend them indefinitely...
```

## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```
H = np.array([[12,11], [7,10]])
J = np.array([[0,1,2],[7,10,4],[3,2,0]])

print(np.linalg.det(H))
print(np.linalg.det(J))
```

## 3.3 Find $H^{-1}$ and then find $J^{-1}$

```
H_inv = np.linalg.inv(H)
print(H_inv)
```

```
J_inv = np.linalg.inv(J)
print(J_inv)
```
