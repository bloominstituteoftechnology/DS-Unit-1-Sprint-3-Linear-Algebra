## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```
mat_H = np.array([[12, 11],
                  [7, 10]])
mat_J = np.array([[0, 1, 2],
                  [7, 10, 4],
                  [3, 2, 0]])

det_H = np.linalg.det(mat_H)
print('Determinant of Matrix H:', det_H)
print('\n')
det_J = np.linalg.det(mat_J)
print('Determinant of Matrix J:', det_J)
```

## 3.3 Find $H^{-1}$ and then find $J^{-1}$

```
inverse_mat_H = np.linalg.inv(mat_H)
print('Inverse of Matrix H:', '\n', inverse_mat_H)
print('\n')
inverse_mat_J = np.linalg.inv(mat_J)
print('Inverse of Matrix J:', '\n', inverse_mat_J)
```
