## 1.3 Scale the vectors you created in 1.1 by $5$, $\pi$, and $-e$ and plot all four vectors (original + 3 scaled vectors) on a graph. What do you notice about these vectors? 

```
from math import e, pi

v4 = v1 * 5
v5 = v1 * pi
v6 = v1 * -e
```

## 1.8 Find $||g||$ and then find $||h||$. Which is longer?

\begin{align}
\vec{g} = \begin{bmatrix} 1 \\ 1 \\ 1 \\ 8 \end{bmatrix}
\qquad
\vec{h} = \begin{bmatrix} 3 \\3 \\ 3 \\ 3 \end{bmatrix}
\end{align}

```
g = np.array([1, 1, 1, 8])
h = np.array([3, 3, 3, 3])

# Norm (magnitude) of g and h:
g_norm = np.sqrt(1**2 + 1**2 + 1**2 + 8**2)
h_norm = np.sqrt(3**2 * 4)

# Check answer:
g_norm_check = np.sqrt(np.dot(g, g))  # ||v||**2 = v·v
assert g_norm == g_norm_check
h_norm_check = np.sqrt(np.dot(h, h))  # ||v||**2 = v·v
assert h_norm == h_norm_check

print('||g|| = ', g_norm)
print('||h|| = ', h_norm, '\n')
if g_norm > h_norm:
  print('Vector g is "longer."')
elif h_norm > g_norm:
  print('Vector h is "longer."')
elif h_norm == g_norm:
  print('Vectors g and h have equal magnitudes.')
```

## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}

\qquad
J= 
\begin{bmatrix}

\end{bmatrix}

```
# Define matrices H and J:
H = np.array([[12, 11], [7, 10]])
J = np.array([[0, 1, 2], 
              [7, 10, 4], 
              [3, 2, 0]])
```

```
# Calculate det(H) and det(J) by hand:
det_H = 12*10 - 7*11
det_J = 0 - (0 - 4*3) + 2 * (7*2 - 3*10)
```

```
# Check answers with NumPy:
assert det_H == round(np.linalg.det(H), 0)
assert det_J == round(np.linalg.det(J), 0)
```
