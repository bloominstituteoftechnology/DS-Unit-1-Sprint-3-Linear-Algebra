## 1.1 Create a two-dimensional vector and plot it on a graph

```
import math 
import numpy as np
import matplotlib.pyplot as plt


two_dim_array = [.5,.5]
two_dim_array = np.array(two_dim_array)
two_dim_array

# Plotting Vectors
plt.arrow(0,0, two_dim_array[0], two_dim_array[1],head_width=.02, head_length=0.02, color ='blue')
plt.xlim(0,1)          
plt.ylim(0,1)
plt.title("2D Vector")
plt.show()
```

## 1.2 Create a three-dimensional vecor and plot it on a graph

```
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

three_dim_array = [0,0,0,.5,.5,.5]
three_dim_array = np.array(three_dim_array)

vector = three_dim_array

X, Y, Z, U, V, W = vector
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(X, Y, Z, U, V, W, length=1)
ax.set_xlim([0, 1])
ax.set_ylim([0, 1])
ax.set_zlim([0, 1])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```
# Determinant of Matrix H
matrix_H = np.array([12,11,7,10]).reshape(2,2)
determinant_H = np.linalg.det(matrix_H)
print(determinant_H)


# Determinant of Matrix J
matrix_J = np.array([0,1,2,7,10,4,3,2,0]).reshape(3,3)
determinant_J = np.linalg.det(matrix_J)
print(determinant_J)
```

## 3.3 Find $H^{-1}$ and then find $J^{-1}$

```
# Matrix H Inverse

inverse_H = np.linalg.inv(matrix_H)
inverse_H
```

```
# Matrix J Inverse

inverse_J = np.linalg.inv(matrix_J)
inverse_J
```

## 3.4 Find $HH^{-1}$ and then find $J^{-1}J$. Is $HH^{-1} == J^{-1}J$? Why or Why not? 

Please ignore Python rounding errors. If necessary, format your output so that it rounds to 5 significant digits (the fifth decimal place).

```
# Product of H and H^(-1)

product_H_and_inverse = np.matmul(matrix_H, inverse_H).round(5)
product_H_and_inverse
```

```
# Product of J and J^(-1)

product_J_and_inverse = np.matmul(inverse_J, matrix_J).round(5)
product_J_and_inverse
```

## **Why?**

They are the same because you're multiplying by the inverse, which gives you the identity matrix. Any matrix, multiplied by it's inverse, will yield you the identity matrix
