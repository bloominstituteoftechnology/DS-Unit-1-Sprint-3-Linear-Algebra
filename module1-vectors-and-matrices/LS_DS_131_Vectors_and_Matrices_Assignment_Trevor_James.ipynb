## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```
# Creat matrices
H = np.array([[12, 11], 
              [7, 10]])

J = np.array([[0,1,2], 
              [7, 10, 4],
              [3, 2, 0]])
```

```
# computing the determinant of both matrices 

print(np.linalg.det(H))
print(np.linalg.det(J))
```

## 3.3 Find $H^{-1}$ and then find $J^{-1}$

```
H_inv = np.linalg.inv(H)
J_inv = np.linalg.inv(J)

print("Inverse of H", H_inv)
print("Inverse of J", J_inv)
```
