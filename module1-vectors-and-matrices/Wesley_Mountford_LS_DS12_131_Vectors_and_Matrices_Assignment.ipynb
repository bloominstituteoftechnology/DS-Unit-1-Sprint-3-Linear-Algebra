```
def vector_dot_product(vec1, vec2):
  if len(vec1) == len(vec2):
    #Going to assume that writing a function that calls np.dot() would be cheating.
    fin_array = np.array([])
    for i in range(len(vec1)):
      fin_array = np.append(fin_array, [(vec1[i] * vec2[i])])
    return fin_array.sum()
  else: #our vectors are not the same length
    print('Function requires two vectors of equal length.')
    return False
```

```
#Testing the above function
vect1 = np.array([3, 2, 3])
vect2 = np.array([1, 4, 4])
print(vector_dot_product(vect1, vect2))
np.dot(vect1, vect2)
```

```
from math import sqrt
def vector_norm(vector):
  #Once again, assuming calling np.linalg.norm() in this function is cheating.
  total = 0
  for i in range(len(vector)):
    total += vector[i]**2
  return sqrt(total)
```
