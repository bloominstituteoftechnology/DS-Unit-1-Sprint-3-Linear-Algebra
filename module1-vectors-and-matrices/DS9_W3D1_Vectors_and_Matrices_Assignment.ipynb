## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```
h = np.array([[12, 11]
             ,[7,  10]
             ]
            )

j = np.array([[0,  1, 2]
             ,[7, 10, 4]
             ,[3,  2, 0]
             ]
            )

print(np.linalg.det(h), '\n'
     ,np.linalg.det(j))
```

```
print(np.linalg.inv(h), '\n\n'
     ,np.linalg.inv(j))
```
