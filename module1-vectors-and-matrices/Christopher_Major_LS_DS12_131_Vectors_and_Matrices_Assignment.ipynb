## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```
H = np.array([[12,11],
              [7,10]])
J=np.array([[0,1,2],
            [7,10,4],
            [3,2,0]])

print('|H| =', np.linalg.det(H))
print('|J| =' , np.linalg.det(J))
```

## 3.3 Find $H^{-1}$ and then find $J^{-1}$

```
H_inv = np.linalg.inv(H)
J_inv = np.linalg.inv(J)
print(H_inv)
print(J_inv)
```
