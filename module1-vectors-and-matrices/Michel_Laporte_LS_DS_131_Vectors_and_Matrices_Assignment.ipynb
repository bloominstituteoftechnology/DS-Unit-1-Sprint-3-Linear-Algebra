## 1.1 Create a two-dimensional vector and plot it on a graph

````
import matplotlib.pyplot as plt
import numpy as np
````

````
# Trial two-dimensional vectors
vector_blue = [3, 1.5]
vector_green = [5, 4]

# Vector plot
plt.arrow(0, 0, vector_blue[0], vector_blue[1], linewidth=3, head_width=0.15, head_length=.25, color = 'b')
plt.arrow(0, 0, vector_green[0], vector_green[1], linewidth=3, head_width=0.15, head_length=.25, color = 'g')
plt.xlim(0, 7)          
plt.ylim(0, 6)
plt.title('Vector Examples')
plt.show()
````

## 1.2 Create a three-dimensional vecor and plot it on a graph

````
from mpl_toolkits.mplot3d import Axes3D

# Trial three-dimensional vectors
vector_red = [5.59, 4.15, 5.40]
vector_black = [2.54, 8.15, 8.45]
vector_yellow = [3.94, 1.54, 1.40]

vectors = [[0, 0, 0, 5.59, 4.15, 5.40],
           [0, 0, 0, 2.54, 8.15, 8.45],
           [0, 0, 0, 3.94, 1.54, 1.40]]

# Vector plot
X, Y, Z, U, V, W = zip(*vectors)
fig = plt.figure()
ax =  fig.add_subplot(111, projection='3d')
ax.quiver(X, Y, Z, U, V, W, length=1)
ax.set_xlim([0, 9])
ax.set_ylim([0, 9])
ax.set_zlim([0, 9])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
````

## 1.3 Scale the vectors you created in 1.1 by $5$, $\pi$, and $-e$ and plot all four vectors (original + 3 scaled vectors) on a graph. What do you notice about these vectors?

## 2.2 Find the following products: CD, AE, and BA. What are the dimensions of the resulting matrices? How does that relate to the dimensions of their factor matrices?

````
a_22 = np.array([[1, 2], [3, 4], [5, 6]])
b_22 = np.array([[2, 4, 6]])
c_22 = np.array([[9, 6, 3], [4, 7, 11]])
d_22 = np.array([[1, 0, 0], [0, 1, 0], [0, 0, 1]])
e_22 = np.array([[1, 3], [5, 7]])

print('The product of CD:')
print(np.matmul(c_22, d_22))
print('')
print('The product of AE:')
print(np.matmul(a_22, e_22))
print('')
print('The product of BA:')
print(np.matmul(b_22, a_22))
````
