## 1.1 Create a two-dimensional vector and plot it on a graph

```
import math
import numpy as np
import matplotlib.pyplot as plt

# Two dimensional graph
# plt.arrow(x, y, dx, dy, **other attributes)
# plt.arrow(x(beginning x location), y(beginning y location), 
# dx(direction of x, or ending x location), dy(direction of y
# or ending y location))

two_d = [4,5]

plt.arrow(0,0, two_d[0], two_d[1], 
          head_width=0.2, head_length = 0.2)
plt.xlim(-1, 7)
plt.ylim(-1, 7)
plt.title("Two Dimensional Plot")
plt.show()
```

## 1.2 Create a three-dimensional vecor and plot it on a graph

```
from mpl_toolkits.mplot3d import Axes3D

three_d = [4,5,6]

'''vectors =  np.array([0,0,0] # X,y,z-axis-location/beginning, 
[4,5,6] # x,y,z-axis direction/ending,); 
length = 1 mean plot exactly what the 3d vector components are
length = 0.5 means plot half of the value of the vector components
length = 2 means plot twice the value of the vector components'''

vectors =  np.array([[0,0,0,4,5,6]])

x,y,z, u, v, w = zip(*vectors)
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.quiver(x,y,z,u,v,w, length = 1)
ax.set_xlim([0,5])
ax.set_ylim([0,6])
ax.set_zlim([0,7])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

## 1.3 Scale the vectors you created in 1.1 by $5$, $\pi$, and $-e$ and plot all four vectors (original + 3 scaled vectors) on a graph. What do you notice about these vectors? 

```
from math import e, pi
print(e)
print(pi)
```

```
two_d = [4,5]
two_d_5 = np.multiply(5, two_d)
two_d_pi = np.multiply(pi, two_d)
two_d_e = np.multiply(-e, two_d)

two_d_plot = plt.arrow(0,0, two_d[0], two_d[1], 
          head_width=0.2, head_length = 2, color='r')
two_d_5_plot = plt.arrow(0,0, two_d_5[0], two_d[1], 
          head_width=0.2, head_length = 2, color='b')
two_d_pi_plot = plt.arrow(0,0, two_d_pi[0], two_d[1], 
          head_width=0.2, head_length = 2, color='y')
two_d_e_plot = plt.arrow(0,0, two_d_e[0], two_d[1], 
          head_width=0.2, head_length = 2)

plt.legend([two_d_plot, two_d_5_plot, two_d_pi_plot, two_d_e_plot ],
           ['Originial', '5-scalar', 'pi-scalar', '-e scalar'])

plt.xlim(-20, 26)
plt.ylim(-2, 10)
plt.title("Two Dimensional Plot with scaled quantities")
plt.show()
```
