## 1.1 Create a two-dimensional vector and plot it on a graph

```
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```
# 2D vector
vec_2d = np.array([2, 5])

# Graph the vector
plt.arrow(0,0,vec_2d[0],vec_2d[1], color='green')
plt.xlim(0, 3)
plt.ylim(0, 5)
plt.title("2-Dimensional Vector")
plt.show()
```

## 1.2 Create a three-dimensional vector and plot it on a graph

```
# 3-dimensional vector
from mpl_toolkits.mplot3d import Axes3D

vectors = np.array([
           [0, 0, 0, 3, 1, -2],
           [0, 0, 0, 1, -2, 2],
           [0, 0, 0, -1, 3, 0]
])
X, Y, Z, U, V, W = zip(*vectors)
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
ax.quiver(X, Y, Z, U, V, W, length=1)
ax.set_xlim(-3, 3)
ax.set_ylim(-3, 3)
ax.set_zlim(-3, 3)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title("3-Dimensional Vector")
plt.show()
```

## 1.8 Find $||g||$ and then find $||h||$. Which is longer?

\begin{align}
\vec{g} = \begin{bmatrix} 1 \\ 1 \\ 1 \\ 8 \end{bmatrix}
\qquad
\vec{h} = \begin{bmatrix} 3 \\3 \\ 3 \\ 3 \end{bmatrix}
\end{align}

```
g = np.array([
              [1, 1, 1, 8]
])

h = np.array([
              [3, 3, 3, 3]
])

g_norm = np.linalg.norm(g)
h_norm = np.linalg.norm(h)

print(f"""
g norm = {g_norm}
h norm = {h_norm}
Therefore, g is longer than h.
""")
```

## 2.1 What are the dimensions of the following matrices? Which of the following can be multiplied together? See if you can find all of the different legal combinations.
\begin{align}
A = \begin{bmatrix}

\end{bmatrix}
\qquad
B = \begin{bmatrix}

\end{bmatrix}
\qquad
C = \begin{bmatrix}
9 & 6 & 3 \\

\end{bmatrix}
\qquad
D = \begin{bmatrix}

0 & 0 & 1
\end{bmatrix}
\qquad
E = \begin{bmatrix}
1 & 3 \\
5 & 7
\end{bmatrix}
\end{align}

```
matrix_A = np.array([
                     [1, 2],
                     [3, 4],
                     [5, 6]
])

matrix_B = np.array([
                     [2, 4, 6]
])

matrix_C = np.array([
                     [9, 6, 3],
                     [4, 7, 11]
])

matrix_D = np.array([
                     [1, 0, 0],
                     [0, 1, 0],
                     [0, 0, 1]
])

matrix_E = np.array([
                     [1, 3],
                     [5, 7]
])

print(f"""
Matrix A {matrix_A.shape}
Matrix B {matrix_B.shape}
Matrix C {matrix_C.shape}
Matrix D {matrix_D.shape}
Matrix E {matrix_E.shape}
""")
```

```
# Multiply matrices A and C
matrix_F = np.matmul(matrix_A, matrix_C)
print(f"""
Matrix F {matrix_F.shape}
{matrix_F}
""")
```

```
# Multiply F (the product of A and C) by D
matrix_G = np.matmul(matrix_F, matrix_D)
print(f"""
Matrix G {matrix_G.shape}
{matrix_G}
""")
```

Matrices A and C can be multiplied, because they have the same shape as the other's transposed shape.

## 2.2 Find the following products: CD, AE, and BA. What are the dimensions of the resulting matrices? How does that relate to the dimensions of their factor matrices?

```
matrix_H = np.matmul(matrix_C, matrix_D)
matrix_I = np.matmul(matrix_A, matrix_E)
matrix_J = np.matmul(matrix_B, matrix_A)
print(f"""
CD = Matrix H {matrix_H.shape}
AE = Matrix I {matrix_I.shape}
BA = Matrix J {matrix_J.shape}
""")
```

CD: the product of matrices with dimensions (2, 3) and (3, 3) has dimensions (2, 3)

AE: the product of matrices with dimensions (3, 2) and (2, 2) has dimensions (3, 2)

BA: the product of matrices with dimensions (1, 3) and (3, 2) has dimensions (1, 2)

Based on these results, the pattern in the products of matrices seems to be to inherit the number of rows of the first factor matrix, and the number of columns of the second factor matrix.

## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```
h = np.array([
              [12, 11],
              [7, 10]
])

j = np.array([
              [0, 1, 2],
              [7, 10, 4],
              [3, 2, 0]
])

# Determinants
det_h = np.linalg.det(h)
det_j = np.linalg.det(j)

print(f"""
|h| = {round(det_h, 3)}
|j| = {round(det_j, 3)}
""")
```

## 3.3 Find $H^{-1}$ and then find $J^{-1}$

```
inv_h = np.linalg.inv(h)
inv_j = np.linalg.inv(j)

print(f"""
h^-1 = {inv_h}
j^-1 = {inv_j}
""")
```
