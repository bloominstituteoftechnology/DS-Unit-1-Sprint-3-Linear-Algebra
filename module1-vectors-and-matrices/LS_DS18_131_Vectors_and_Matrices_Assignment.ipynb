## 2.1 What are the dimensions of the following matrices? Which of the following can be multiplied together? See if you can find all of the different legal combinations.
\begin{align}
A = \begin{bmatrix}
1 & 2 \\

\end{bmatrix}
\qquad
B = \begin{bmatrix}

\end{bmatrix}
\qquad

\qquad

\qquad
E = \begin{bmatrix}
1 & 3 \\
5 & 7

\end{align}

```
matrix_a = np.array([[1,2],
                     [3,4],
                     [5,6]])

matrix_b = np.array([2,4,6])

matrix_c = np.array([[9,6,3],
                     [4,7,11]])

matrix_d = np.array([[1,0,0],
                     [0,1,0],
                     [0,0,1]])

matrix_e = np.array([[1,3],
                     [5,7]])

print('Matrix A Shape\n', matrix_a.shape)
print('Matrix B Shape\n', matrix_b.shape)
print('Matrix C Shape\n', matrix_c.shape)
print('Matrix D Shape\n', matrix_d.shape)
print('Matrix E Shape\n', matrix_e.shape, '\n')

print('Matrix A x Matrix C\n', np.matmul(matrix_a, matrix_c), '\n')
print('Matrix A x Matrix E\n', np.matmul(matrix_a, matrix_e), '\n')

print('Matrix B x Matrix A\n', np.matmul(matrix_b, matrix_a), '\n')
print('Matrix B x Matrix D\n', np.matmul(matrix_b, matrix_d), '\n')

print('Matrix C x Matrix A\n', np.matmul(matrix_c, matrix_a), '\n')
print('Matrix C x Matrix D\n', np.matmul(matrix_c, matrix_d), '\n')

print('Matrix D x Matrix A\n', np.matmul(matrix_d, matrix_a), '\n')

print('Matrix E x Matrix C\n', np.matmul(matrix_e, matrix_c), '\n')
```

## 2.2 Find the following products: CD, AE, and BA. What are the dimensions of the resulting matrices? How does that relate to the dimensions of their factor matrices?

```
print('Matrix C x Matrix D\n', np.matmul(matrix_c, matrix_d))
print('Combined Matrix Shape', np.matmul(matrix_c, matrix_d).shape)
print('Matrix C Shape', matrix_c.shape)
print('Matrix D Shape', matrix_d.shape, '\n')

print('Matrix A x Matrix E\n', np.matmul(matrix_a, matrix_e))
print('Combined Matrix Shape', np.matmul(matrix_a, matrix_e).shape)
print('Matrix A Shape', matrix_a.shape)
print('Matrix E Shape', matrix_e.shape, '\n')

print('Matrix B x Matrix A\n', np.matmul(matrix_b, matrix_a))
print('Combined Matrix Shape', np.matmul(matrix_b, matrix_a).shape)
print('Matrix B Shape', matrix_b.shape)
print('Matrix A Shape', matrix_a.shape, '\n')

# The dimensions of the resulting matrices are equal to the leftover
# outer dimensions
# The inner dimensions are used to see if a matrix multiplication operation
# would be valid
```

## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```
matrix_h = np.array([[12, 11],
                     [7, 10]])

matrix_j = np.array([[0,1,2],
                     [7,10,4],
                     [3,2,0]])

print("Determinant of H:", np.linalg.det(matrix_h))
print("Determinant of J:", np.linalg.det(matrix_j))
```

## 3.3 Find $H^{-1}$ and then find $J^{-1}$

```
print("H^-1:\n", np.linalg.inv(matrix_h))
print("J^-1:\n", np.linalg.inv(matrix_j))
```
