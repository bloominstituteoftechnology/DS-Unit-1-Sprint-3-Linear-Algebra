## 1.6 Find $c \cdot d$

\begin{align}
\vec{c} = \begin{bmatrix}7 & 22 & 4 & 16\end{bmatrix}
\qquad
\vec{d} = \begin{bmatrix}12 & 6 & 2 & 9\end{bmatrix}
\end{align}


```
def dot_product(x, y):
  return (np.array(x) * np.array(y)).sum()

c = [7, 22, 4,16]
d = [12, 6, 2, 9]

dot_product(c,d)
```

## 1.8 Find $||g||$ and then find $||h||$. Which is longer?

\begin{align}
\vec{g} = \begin{bmatrix} 1 \\ 1 \\ 1 \\ 8 \end{bmatrix}
\qquad
\vec{h} = \begin{bmatrix} 3 \\3 \\ 3 \\ 3 \end{bmatrix}
\end{align}

```
def vec_length(x):
  return math.sqrt((np.array(x) * np.array(x)).sum())

g = [1,1,1,8]
h = [3,3,3,3]


print(vec_length(g))
print(vec_length(h))
```

## 2.2 Find the following products: CD, AE, and BA. What are the dimensions of the resulting matrices? How does that relate to the dimensions of their factor matrices?

```
C = np.array([[9,6,3],[4,7,11]])
D = np.array([[1,0,0],[0,1,0], [0,0,1]])

# Dimension of the resulting matrix 
print("Dimension of the resulting matrix is", (np.matmul(C,D)).shape)

np.matmul(C,D)
```

```
A = np.array([[1,2], [3,4], [5,6]])
E = np.array([[1,3], [5,7]])

# Dimension of the resulting matrix 
print("Dimension of the resulting matrix is", np.matmul(A,E).shape)

np.matmul(A,E)
```

## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}

\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\

\end{bmatrix}
\end{align}


```
H = np.array([[12,11],[7,10]])

det_H = (12*10) -(7*11)
print("|H| is equal to ", det_H)

J = np.array([[0,1,2], [7,10,4], [3,2,0]])
det_J = 0 - 1*(0*7 - 3*4) + 2*(7*2-10*3)
print("|J| is equal to ", det_J)
```
