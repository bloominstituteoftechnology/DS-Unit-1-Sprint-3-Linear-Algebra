## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```
h = np.array([[12,11], [7,10]])

j = np.array([[0,1,2], [7,10,4], [3,2,0]])

hd = np.linalg.det(h)

jd = np.linalg.det(j)

print(hd)

print(jd)
```

## 3.3 Find $H^{-1}$ and then find $J^{-1}$

```
hi = np.linalg.inv(h)

ji = np.linalg.inv(j)

print(hi.shape)
print(hi)

print(ji.shape)
print(ji)
```

## 3.4 Find $HH^{-1}$ and then find $J^{-1}J$. Is $HH^{-1} == J^{-1}J$? Why or Why not? 

Please ignore Python rounding errors. If necessary, format your output so that it rounds to 5 significant digits (the fifth decimal place).

HH^-1 and J^-1J are not equal to each other, but they are both index matrices of their respective factors. This is because multiplying any matrix by the inverse of that same matrix will give you the index of that matrix. Any time you multiply something by it's own reciprocal , you get 1, which is why this produces matrix indices.

```
hh = np.matmul(h,hi)

print(hh.round(5))

jj = np.matmul(ji,j)

print(jj.round(5))
```
