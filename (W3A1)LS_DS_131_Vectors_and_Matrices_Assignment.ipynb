```
#Function for calculating the dot product of equal length
def dot_func(vec_1,vec_2):
  test_list = []
  if len(vec_1) == len(vec_2):
    r = len(vec_1)
    for i in range(r):
      test_list.append(vec_1[i]*vec_2[i])
    return test_list
  else:
    return print('Sorry the two vectors are not of equal length')

a = [1,2,3]
b = [5,6,7,]
dot_func(a,b)
```

```
#Function for calculating the norm of any vector
def norm_func(vector):
  test_var = 0
  for i in vector:
    test_var = test_var + (i**2)
  return test_var**(1/2)

a = [1,2,3]
norm_func(a)
```
