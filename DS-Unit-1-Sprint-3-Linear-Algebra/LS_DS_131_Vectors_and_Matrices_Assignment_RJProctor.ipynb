```
coffee = [7,5,9]

vector = np.array([[0, 0, 0, coffee[0], coffee[1], coffee[2]]])

X, Y, Z, U, V, W = zip(*vector)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(X, Y, Z, U, V, W, length=1)
ax.set_title(label="3D graphs for 3D vectors.", color='red')
ax.set_xlim([0, 10])
ax.set_ylim([0, 10])
ax.set_zlim([0, 10])
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

plt.show()
```

## 1.3 Scale the vectors you created in 1.1 by $5$, $\pi$, and $-e$ and plot all four vectors (original + 3 scaled vectors) on a graph. What do you notice about these vectors? 

```
# Start with a simple vector
apples = [15, 4]

# Then multiply it by a scalar
# (5, pi, and -e are the scalar vectors below, single number)
green = np.multiply(5, apples)

red = np.multiply(math.pi, apples)

orange = np.multiply(-e, apples)

# Plot the Scaled Vectors
plt.arrow(0,0, apples[0], apples[1],head_width=.05, head_length=0.05, color ='blue')
plt.arrow(0,0, red[0], red[1],head_width=.05, head_length=0.05, color ='red')
plt.arrow(0,0, green[0], green[1],head_width=.05, head_length=0.05, color ='green')
plt.arrow(0,0, orange[0], orange[1],head_width=.05, head_length=0.05, color ='orange')
plt.xlim(-40,20)          
plt.ylim(-15,10)
plt.title("Scaled Vectors")
plt.show()
```

## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\

\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```
j = pd.DataFrame({'a': [0,7,3], 'b': [1,10,2], 'c': [2,4,0]})

print(j.shape)
j.head()
```

```
# J^T
print(j.T.shape)
j.T.values
```

```
# |H| = a*d  b*c = (12*10)-(11*7)
# |J| = a*e*i + b*f*g + c*d*h - c*e*g - b*d*i - a*f*h = (0*10*0) + (1*3*2) + (2*7*2) - (2*10*3) - (1*7*0) - (0*4*2)
H = np.array([[12,7],[11,10]])
J = np.array([[0,1,2],[7,10,4],[3,2,0]])
```
