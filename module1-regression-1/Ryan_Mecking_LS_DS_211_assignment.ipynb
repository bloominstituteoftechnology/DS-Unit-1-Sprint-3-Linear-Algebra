```python
import sys

# If you're on Colab:
if 'google.colab' in sys.modules:
    DATA_PATH = 'https://raw.githubusercontent.com/LambdaSchool/DS-Unit-2-Applied-Modeling/master/data/'

# If you're working locally:
else:
    DATA_PATH = '../data/'
    
# Ignore this Numpy warning when using Plotly Express:
# FutureWarning: Method .ptp is deprecated and will be removed in a future version. Use numpy.ptp instead.
import warnings
warnings.filterwarnings(action='ignore', category=FutureWarning, module='numpy')
```

```python
# Read New York City apartment rental listing data
import pandas as pd
df = pd.read_csv(DATA_PATH+'apartments/renthop-nyc.csv')
assert df.shape == (49352, 34)
```

```python
# Remove outliers: 
# the most extreme 1% prices,
# the most extreme .1% latitudes, &
# the most extreme .1% longitudes
df = df[(df['price'] >= 1375) & (df['price'] <= 15500) & 
        (df['latitude'] >=40.57) & (df['latitude'] < 40.99) &
        (df['longitude'] >= -74.1) & (df['longitude'] <= -73.38)]
```

```python
df.head()
```

```python
#  Look at the data. Choose a feature, and plot its relationship with the target.
import plotly.express as px
px.scatter(df, x='price', y='bedrooms', trendline='ols')
```

```python
# Use scikit-learn for linear regression with one feature.
from sklearn.linear_model import LinearRegression
# Instantiate this class(makes it easier to work with)
model = LinearRegression()
```

```python
# Creating x features matrix and y target vector
features = ['bedrooms']
target = 'price'
X_train = df[features]
y_train = df[target]
```

```python
# Fitting the moddle
model.fit(X_train, y_train)
```

```python
#Applying the data
bedroom_numbers = 3
X_test = [[bedroom_numbers]]
y_pred = model.predict(X_test)
```

```python
#The prediction is 
y_pred
```

```python
#Define a function to make new predictions and explain the model coefficient.
def predict(bedrooms):
  y_pred = model.predict([[bedrooms]])
  estimate = y_pred[0]
  coefficient = model.coef_[0]
  result = f'${estimate:,.0f} estimated price for {bedrooms:,.0f} bedroom condos.'
  explanation = f'In this linear regression, each additional bedroom adds ${coefficient:,.0f}.'
  return result + '\n' + explanation

print(predict(3))
```

```python
from ipywidgets import interact
interact(predict, bedrooms=(0,250));
```

```python
# Do linear regression with two or more features.
features = (['bedrooms', 'bathrooms'])
target = 'price'
X_train = df[features]
y_train = df[target]
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
room_numbers = [3, 25]
X_test = [room_numbers]
y_pred = model.predict(X_test)
y_pred
```

```python
def predict(x1):
  y_pred = model.predict([x1])
  estimate = y_pred[0]
  coefficient = model.coef_[0]
  result = f'${estimate:,.0f} estimated price for {x1[0]:,.0f} bedroom and {x1[1]:,.0f} bathroom condos.'
  explanation = f'In this linear regression, each additional bedroom adds ${coefficient:,.0f}.'
  return result + '\n' + explanation

print(predict([3, 25]))
```
