Write a loop that iterates through all the columns in the dataframe and returns the pearson coefficient for all numerical values. 

```python
import scipy.stats as st

for i in bikes.select_dtypes([int, float]):
    print(f" {i}: {st.pearsonr(bikes[i], bikes['cnt'])[0]}")
```

```python
def model_maker(feature):
    from statsmodels.formula.api import ols
    model = ols(f'cnt ~ {feature}', bikes ).fit()
    print(model.params[0], model.params[1], st.pearsonr(bikes[i], bikes['cnt'])[1]) 
```
