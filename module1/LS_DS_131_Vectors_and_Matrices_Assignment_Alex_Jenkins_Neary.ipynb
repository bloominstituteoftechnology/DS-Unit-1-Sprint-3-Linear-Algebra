## 1.1 Create a two-dimensional vector and plot it on a graph

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
def plot_2d_vector(vector, coloration):
  plt.arrow(0,0, vector[0], vector[1], head_width = .45, color = coloration)
```

```
vector = [2, 5]
plot_2d_vector(vector, "blue")
plt.xlim(0, 10)
plt.ylim(0, 10)
plt.title("Vector Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

## 1.2 Create a three-dimensional vecor and plot it on a graph

```
from mpl_toolkits import mplot3d
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
vector2 = [[0, 0, 0, 4, 6, 2]]
X, Y, Z, U, V, W = zip(*vector2)
ax.quiver(X, Y, Z, U, V, W)
ax.set_xlim([0, 10])
ax.set_ylim([0, 10])
ax.set_zlim([0, 10])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.title("Vector Plot")
plt.show()
```

## 1.3 Scale the vectors you created in 1.1 by $5$, $\pi$, and $-e$ and plot all four vectors (original + 3 scaled vectors) on a graph. What do you notice about these vectors? 

```
from math import e, pi
print(e)
print(pi)
```

## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```
H = np.array([[12, 11], [7, 10]])
print(np.linalg.det(H))
J = np.array([[0, 1, 2], [7, 10, 4], [3, 2, 0]])
print(np.linalg.det(J))
```

## 3.3 Find $H^{-1}$ and then find $J^{-1}$

```
H_inv = np.linalg.inv(H)
J_inv = np.linalg.inv(J)
print(H_inv)
print(J_inv)
```

## 3.4 Find $HH^{-1}$ and then find $J^{-1}J$. Is $HH^{-1} == J^{-1}J$? Why or Why not? 

Please ignore Python rounding errors. If necessary, format your output so that it rounds to 5 significant digits (the fifth decimal place).

```
HH_inv = np.matmul(H, H_inv)
J_invJ = np.matmul(J_inv, J)
print(np.round(HH_inv, 2))
print(np.round(J_invJ, 2))
HH_inv == J_invJ
```

The two identity matrices are not equal; they are different dimensions (2x2, 3x3).
