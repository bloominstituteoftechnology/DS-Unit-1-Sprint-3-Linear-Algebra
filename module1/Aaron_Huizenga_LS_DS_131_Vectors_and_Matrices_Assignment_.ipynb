## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```
H = np.array([[12, 11],
             [7, 10]])
J= np.array([[0, 1, 2],
            [7, 10, 4],
            [3, 2, 0]])

print(np.linalg.det(H))
print(np.linalg.det(J))
```
