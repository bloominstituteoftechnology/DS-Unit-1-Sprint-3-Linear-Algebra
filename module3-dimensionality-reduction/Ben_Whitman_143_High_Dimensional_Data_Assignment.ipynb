## 4.2 Verify that your transformation matrices are correct by choosing an input matrix and calculating the result both via the traditional functions above and also via vector-matrix multiplication.

```
import numpy as np
```

```
input_1 = np.matrix([[2],[3]])
print('input:\n', input_1)

T_p = np.matrix([[1,3],[-1,2]])
print('transform matrix:\n', T_p)

print('result:\n', T_p * input_1)
```

Result of multiplying input by transform matrix matches result of inputting [2,3] by hand

```
input_2 = np.matrix([[3],[4],[5]])
print('input:\n', input_2)

T_q = np.matrix([[4,1,2],[-1,2,3],[5,1,-2]])
print('transform matrix:\n', T_q)

print('result:\n', T_q * input_2)
```

Result of multiplying input by transform matrix matches result of inputting [3,4,5] by hand.

# Eigenvalues and Eigenvectors

## 5.1 In your own words, give an explanation for the intuition behind eigenvalues and eigenvectors.

An eigenvector is a part of a vector space that, given a certain transformation, does not transform. It's an axis around which the vectors transform, but it does not itself move.

An eigenvalue is a scalar that represents how much its corresponding eigenvalue is scaled by its transformation.

# The Curse of Dimensionality

## 6.1 What are some of the challenges of working with high dimensional spaces?

You run the risk of overfitting your line to your data.

Analyzing the data becomes harder and harder by an order of magnitude with each new dimension.

Harder to find meaningful relationships within the data.

## 6.2 What is the rule of thumb for how many observations you should have compared to parameters in your model?

You should have at least 5 times as many observations as you do parameters. At smaller numbers of parameters, you should aim to have observations = parameters^2

The code below will read in the dataset and perform the one-hot encoding of the categorical variables. Start adding your PCA code at the bottom of the provided code.

```
import os.path
from urllib.request import urlopen
from zipfile import ZipFile
from io import BytesIO
import pandas as pd

# Read National Data 
national_url = 'https://www.huduser.gov/portal/datasets/hads/hads2013n_ASCII.zip'
national_file = 'thads2013n.txt'

if os.path.exists(national_file):
    national = pd.read_csv(national_file)
else: 
    z_national = urlopen(national_url)
    zip_national = ZipFile(BytesIO(z_national.read())).extract(national_file)
    national = pd.read_csv(zip_national)
```

```
# Look at first 5 rows
print(national.shape)
national.head()
```

```
# Look at datatypes
# a lot of object datatypes even though they look to be strings of numbers.
national.dtypes
```

```
list(national.columns)
```

```
# check for null values
national.isnull().sum().any()
```

```
# check for number of categorical vs numeric columns
cat_cols = national.columns[national.dtypes=='object']
num_cols = national.columns[national.dtypes!='object']

print(f'{len(cat_cols)} categorical columns')
print(f'{len(num_cols)} numerical columns')
```

```
national_processed = national.copy()

# Get One-Hot-Encoding of categorical variables
national_processed = pd.get_dummies(national_processed)
print(national_processed.shape)
national_processed.head()
```

```
# Create random subset of dataframe for processing
national_processed = national_processed.sample(frac=0.01)
```

```
national_processed.shape
```

```
# Isolate dependent variable
Y = national_processed['ZSMHC']
```

```
Y.head()
```

```
# Gather independent variables
X = national_processed.drop('ZSMHC', axis=1)
```

```
X.head()
```

```
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```
scaler = StandardScaler()
Z = scaler.fit_transform(X)
```

```
pca = PCA(n_components = 2)
pca.fit(Z)
```

```
df_pca = pd.DataFrame(pca.components_.T)
```

```
df_pca
```

```
plt.scatter(df_pca[0], df_pca[1])
```
