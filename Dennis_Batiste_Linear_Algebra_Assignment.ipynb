## 1.2 Create a three-dimensional vecor and plot it on a graph

```
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

vector = np.array ([0,0,.6,.6,.8,.10])
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
X, Y, Z, U, V, W = [*vector]
ax.quiver(X, Y, Z, U, V, W)
ax.set_xlim([0,1])
ax.set_ylim([0,1])
ax.set_zlim([0,1])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

```
a = [[1, 2], [3, 4], [5, 6]]
b = [2, 4, 6]
c = [[9, 6, 3], [4, 7, 11]]
d = [[1, 0, 0], [0, 1, 0], [0, 0, 1]]
e = [[1, 3], [5, 7]]
print('CD = \n', np.dot(c, d)) # dimension of resulting matrix: 2x3
print('AE = \n', np.dot(a, e)) # dimension of resulting matrix: 3x2
print('BA = \n', np.dot(b, a)) # dimension of resulting matrix: 1x2
# dimensions of resulting matrices are the rows of the right matrix by the columns of the left.
```
